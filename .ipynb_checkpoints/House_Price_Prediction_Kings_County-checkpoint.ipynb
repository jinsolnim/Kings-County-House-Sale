### 3. First Simple Model

```python
fsm_df = df[["sqft_living", "price"]].copy()
```

```python
fsm = ols(formula="price ~ sqft_living", data=fsm_df)
fsm_results = fsm.fit()
```

```python
fsm_results.summary()
```

### Evaluation
The 49.3% represents percentage of variation in dependent that is explained by independent variable. It means sqft living only explains 49.3% of the variance in the price.
The price is $279.93 for one sqft_living.

#### Linearity

```python
rainbow_statistic, rainbow_p_value = linear_rainbow(fsm_results)
print("Rainbow statistic:", rainbow_statistic)
print("Rainbow p-value:", rainbow_p_value)
```

The null hypothesis is that the model is linearly predicted by the features, alternative hypothesis is that it is not. Thus returning a low p-value means that the current model violates the linearity assumption.

#### Normality
Currently **Jarque-Bera (JB)** is 516072.474.

The null hypothesis is that the residuals are normally distributed, alternative hypothesis is that they are not. Thus returning a low p-value means that the current model violates the normality assumption.

#### Homoscadasticity

Linear regression assumes that the variance of the dependent variable is homogeneous across different value of the independent variable(s).  We can visualize this by looking at the predicted price vs. the residuals.

```python
y = fsm_df["price"]
y_hat = fsm_results.predict()
```

```python
fig2, ax2 = plt.subplots(figsize =(10,6))
ax2.set(xlabel="Predicted Price",
        ylabel="Residuals (Predicted - Actual Price)")
ax2.scatter(x=y_hat, y=y_hat-y, color="blue", alpha=0.2);
```

```python
#running Breusch-Pagan test (statistical test)
lm, lm_p_value, fvalue, f_p_value = het_breuschpagan(y-y_hat, fsm_df[["sqft_living"]])
print("Lagrange Multiplier p-value:", lm_p_value)
print("F-statistic p-value:", f_p_value)
```

since p-value is < 0.05, null hypothesis is rejected. Therefore current model violates the homoscedasticity assumption.

#### Independence
current model only has one independent variable.

***seeing if log transformation of the price would help the model***

```python
logprice = pd.DataFrame([])
logprice['log_price'] = np.log(fsm_df['price'])
logprice['sqft_living'] = fsm_df['sqft_living']
```

```python
smd = logprice.copy()
```

```python
smd = ols(formula="log_price ~ sqft_living", data=smd)
smd_results = smd.fit()
```

```python
smd_results.summary()
```

```python
#checking linearity
rainbow_statistic, rainbow_p_value = linear_rainbow(smd_results)
print("Rainbow statistic:", rainbow_statistic)
print("Rainbow p-value:", rainbow_p_value)
```

still violating linearity assumption but improved

Normality became much better compare to first simple model **Jarque-Bera (JB)** is 3.675

#### Homoscadasticity

```python
y1 = logprice["log_price"]
y_hat1 = smd_results.predict()
```

```python
fig6, ax6 = plt.subplots(figsize =(10,6))
ax6.set(xlabel="Predicted Price",
        ylabel="Residuals (Predicted - Actual Price)")
ax6.scatter(x=y_hat1, y=y_hat1-y1, color="blue", alpha=0.2);
```

```python
lm, lm_p_value, fvalue, f_p_value = het_breuschpagan(y1-y_hat1, logprice[["sqft_living"]])
print("Lagrange Multiplier p-value:", lm_p_value)
print("F-statistic p-value:", f_p_value)
```

since p-value is < 0.05, null hypothesis is rejected. Therefore current model violates the homoscedasticity assumption.

**Independence** is still not applicable.

### 4. Adding features to the model

```python
df['log_price'] = np.log2(df['price'])
```

```python
outcome = 'log_price'
x_columns = ['sqft_living', 'bathrooms', 'sqft_lot', 'sqft_above', 'sqft_living15', 'sqft_lot15', 'yr_built']
predictors = '+'.join(x_columns)
formula = outcome + '~' + predictors
model = ols(formula=formula, data=df).fit()
model.summary()
```

```python
pd.plotting.scatter_matrix(df[x_columns], figsize=(10,12));
```

### Evaluation
Adding variables increased the r-squared from 0.492 to 0.549.

#### Linearity

```python
rainbow_statistic, rainbow_p_value = linear_rainbow(model)
print("Rainbow statistic:", rainbow_statistic)
print("Rainbow p-value:", rainbow_p_value)
```

We are still violating the linearity assumption since p-value is still low.

#### Normality
The **Jarque-Bera (JB)** output has been increased from 516072.474 to 709226.194. Therefore, We are still violating the normality assumption.

#### Homoscadasticity

```python
df2 = ['log_price', 'sqft_living', 'bathrooms', 'sqft_lot', 'sqft_above', 'sqft_living15', 'sqft_lot15', 'yr_built']
```

```python
df2 = df.astype({'bathrooms': 'int'})
```

```python
y = df2["log_price"]
y_hat = model.predict()
```

```python
fig4, ax4 = plt.subplots(figsize=(10,10))
ax4.set(xlabel="Predicted Price",
        ylabel="Residuals (Predicted - Actual Price)")
ax4.scatter(x=y_hat, y=y_hat-y, color="blue", alpha=0.2);
```

```python
lm, lm_p_value, fvalue, f_p_value = het_breuschpagan(y-y_hat, df2[["sqft_living", "bathrooms", "sqft_lot", "sqft_above", "sqft_living15", "sqft_lot15", "yr_built"]])
print("Lagrange Multiplier p-value:", lm_p_value)
print("F-statistic p-value:", f_p_value)
```

since p-value is < 0.05, null hypothesis is rejected. Therefore current model violates the homoscedasticity assumption.

#### Independence

```python
#creating variance inflation factor to measure multicollinearity
rows = df2[["sqft_living", "bathrooms", "sqft_lot", "sqft_above", "sqft_living15", "sqft_lot15", "yr_built"]].values

vif_df = pd.DataFrame()
vif_df["VIF"] = [variance_inflation_factor(rows, i) for i in range(7)]
vif_df["feature"] = ["sqft_living", "bathrooms", "sqft_lot", "sqft_above", "sqft_living15", "sqft_lot15", "yr_built"]

vif_df
```

It seems like the second model violating the independence.

```python
#creating new dataframe
pre = df.loc[:, x_columns]
pre.head()
```

```python
pre.corr()
```

```python
#Return True for positive or negative correlations that are bigger than 0.75 in the correlation 
abs(pre.corr()) > 0.75
```

```python
#Now, include stack and zip to create a more robust solution that will return the variable pairs from the correlationthat have correlations over .75, but less than 1.
ind = pre.corr().abs().stack().reset_index().sort_values(0, ascending=False)

ind['pairs'] = list(zip(ind.level_0, ind.level_1))

ind.set_index(['pairs'], inplace = True)

ind.drop(columns=['level_1', 'level_0'], inplace = True)

# cc for correlation coefficient
ind.columns = ['cc']

ind.drop_duplicates(inplace=True)

ind[(ind.cc>.75) & (ind.cc<1)]
```

sqft_living seems highly correlated

## Summary

My first simple model started with one independent variable ‘sqft_living’. First model had an r-squared of 0.492. This model violated the linearity , normality, and homoscedasticity assumptions of linear regression (p<0.001 for all tests). Since it was the only feature, it met the independence assumption. 

The final model had 14 input features : ‘sqft_living’, ‘bathrooms’, ‘sqft_lot’, ‘sqft_above’, ‘sqft_living15’, ‘sqft_lot15’, ‘yr_built’, ‘waterfront’, ‘bedrooms’, ‘floors’, ‘view’, ‘condition’, ‘grade’, ‘zipcode’. The final model had improved r-squared from 0.492(first simple model) to 0.655. However, this model still violated linearity assumption (p>0.05), normality (p<0.001) and homoscedasticity (p<0.001) assumptions. It also violated the independence assumption.
